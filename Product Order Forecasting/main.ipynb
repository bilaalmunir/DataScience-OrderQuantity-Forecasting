```python
import pandas as pd
import matplotlib as mpl
import os
```

```python
file_paths={
16 : 'Data/Shipments and Orders 2016.csv',
17 : 'Data/Shipments and Orders 2017.csv',
18 : 'Data/Shipments and Orders 2018.csv',
19 : 'Data/Shipments and Orders 2019.csv',
20 : 'Data/Shipments and Orders 2020.csv',
21 : 'Data/Shipments and Orders 2021.csv',
22 : 'Data/Shipments and Orders 2022.csv',
23 : 'Data/Shipments and Orders 2023.csv',
24 : 'Data/Shipments and Orders 2024 may.csv',
}
```

```python
years = range(16,24)
dataframes = {}

for year,path in file_paths.items():
    try:
        dataframes[year] = pd.read_csv(path)
    except FileNotFoundError:
        print(f"file not found : {path}")
    
```

```python
for item in dataframes.items():
    print(item)
```

```python
from pathlib import Path
new_file_path = Path('Data/concatCSV.csv')

if not new_file_path.exists() : 
    new_file_path.touch()

combined_df = pd.concat(dataframes.values(), ignore_index=True)
combined_df.to_csv(new_file_path)
    
```
